```python
import pandas as pd
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
stock_04_log1 = pd.read_csv("../logs/20240106.csv")
stock_04_log1.head(10)
```

```python
df_rpm = stock_04_log1['Engine Speed (rpm)']
stock_04_log1.drop(['Time (msec)','Engine Speed (rpm)'], axis=1, inplace=True)
```

```python
stock_04_log1[['Throttle Opening Angle (%)','Mass Airflow (g/s)','A/F Sensor #1 (AFR)',
               'A/F Correction #1 (%)','A/F Learning #1 (%)','Injector Duty Cycle (%)',
              'Engine Load (Calculated) (g/rev)','Manifold Relative Pressure (Corrected) (psi)',
               'Primary Wastegate Duty Cycle (%)','Ignition Total Timing (degrees)',
              'IAM* (raw ecu value)','Fine Learning Knock Correction* (degrees)',
               'Feedback Knock Correction* (degrees)','Knock Correction Advance (degrees)']]
```

```python
# use Plotly express, subplots, and graph objects to create dual y-axis plot
fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(
    go.Scatter(y=stock_04_log1['Throttle Opening Angle (%)'], name="Throttle", line=dict(color='orange')),
    secondary_y=False
)
fig.add_trace(
    go.Scatter(y=stock_04_log1['Mass Airflow (g/s)'], name="Airflow (g/s)", line=dict(color='skyblue')),
    secondary_y=False
)
fig.add_trace(
    go.Scatter(y=stock_04_log1['A/F Sensor #1 (AFR)'], name="AFR", line=dict(color='navy')),
    secondary_y=False
)
fig.add_trace(
    go.Scatter(y=stock_04_log1['Injector Duty Cycle (%)'], name="Inj Duty", line=dict(color='pink')),
    secondary_y=False
)
fig.add_trace(
    go.Scatter(y=stock_04_log1['Manifold Relative Pressure (Corrected) (psi)'], name="Boost", line=dict(color='gold')),
    secondary_y=False
)
fig.add_trace(
    go.Scatter(y=stock_04_log1['Primary Wastegate Duty Cycle (%)'], name="WG Duty", line=dict(color='purple')),
    secondary_y=False
)
fig.add_trace(
    go.Scatter(y=stock_04_log1['Ignition Total Timing (degrees)'], name="Total Ign Timing", line=dict(color='limegreen')),
    secondary_y=False
)
fig.add_trace(
    go.Scatter(y=stock_04_log1['IAM* (raw ecu value)'], name="IAM", line=dict(color='forestgreen')),
    secondary_y=False
)
fig.add_trace(
    go.Scatter(y=stock_04_log1['Feedback Knock Correction* (degrees)'], name="FBKC", line=dict(color='black')),
    secondary_y=False
)
fig.add_trace(
    go.Scatter(y=stock_04_log1['Fine Learning Knock Correction* (degrees)'], name="FLKC", line=dict(color='brown')),
    secondary_y=False
)
fig.add_trace(
    go.Scatter(y=stock_04_log1['Knock Correction Advance (degrees)'], name="Knock Corr Adv", line=dict(color='grey')),
    secondary_y=False
)

# Add the secondary Y trace
fig.add_trace(
    go.Scatter(y=df_rpm, name="RPM", line=dict(color='red')),
    secondary_y=True
)

# Format
fig.update_layout(
    title_text="MY 04 WRX Stock 3rd Gear Pull",
    width=800,
    height=600
)

fig.update_layout(showlegend=True)

fig.update_layout(hovermode="x unified")

fig.layout.yaxis1.update(showticklabels=False)

# Set y-axis title
fig.update_yaxes(title_text="<b>Engine RPM</b>", secondary_y=True)

fig.show()
```
